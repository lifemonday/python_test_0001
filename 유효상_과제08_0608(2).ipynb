```python
contact = pd.read_csv('c:/AIPython_2/contacts.csv', encoding='euc-kr')
contact
```

```python
contact.iloc[:, 1]=contact.iloc[:,1].replace('휴직', np.nan)
contact
```

```python
def remove(x):
    return x.replace(' ','')
```

```python
contact.iloc[:,0]=contact.iloc[:,0].apply(remove)
```

```python
contact.rename(columns={'사무실':'사무실 전화번호'}, inplace=True)
```

```python
contact.iloc[:,1]='(02)870-'+contact.iloc[:,1]
contact
```

```python
contact.xs(42)['사무실 전화번호']='043-249-0121(오송)'
contact
```

```python
work_iter = (set(x.split(',')) for x in contact['담당업무'])
display(work_iter)
work = sorted(set.union(*work_iter))
work
```

```python
dummies = DataFrame(np.zeros((len(contact), len(work))), columns=work)
```

```python
dummies.head()
```

```python
for i, gen in enumerate(contact['담당업무']):
    dummies.iloc[i].loc[gen.split(',')] = 1
```

```python
contact_work = contact.join(dummies.add_prefix('Work_'))
contact_work.iloc[0].head(10)
```
