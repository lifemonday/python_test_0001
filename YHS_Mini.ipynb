```python
import operator
```

```python
flist1=zip(sisullist,alllist)
flist2=zip(sisullist,menlist)
flist3=zip(sisullist,womenlist)
flist1.sort(key=operator.itemgetter(0))
flist2.sort(key=operator.itemgetter(0)) 
flist3.sort(key=operator.itemgetter(0))
```

```python
sisullist1=[i[0] for i in flist1]
alllist1=[i[1] for i in flist1]
```

```python
sisullist2=[i[0] for i in flist2]
menlist2=[i[1] for i in flist2]
```

```python
sisullist3=[i[0] for i in flist3]
womenlist3=[i[1] for i in flist3]
```

```python
fig, axes = plt.subplots(3, 1,sharex=True)
```

```python
axes[0].plot(sisullist1,alllist1,'g', label='all')
axes[1].plot(sisullist2,menlist2,'b', label='men')
axes[2].plot(sisullist3,womenlist3,'r', label='women')
axes[0].legend(loc='best')
axes[1].legend(loc='best')
axes[2].legend(loc='best')
axes[2].set_xlabel('sesul-->')
```
